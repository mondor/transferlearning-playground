```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Efficientnet-hamiltonisland-status.csv', names=["id", "image", "published", "disabled"], header=None)
df.head()
```

```python
# make data folders
import os

base_dir = './hamiltonisland-data'
os.makedirs(base_dir, exist_ok=True)

for folder in ['train', 'validation', 'test']:
    top_path = os.path.join(base_dir, folder)
    os.makedirs(top_path, exist_ok=True)
    
    for cat in ['published', 'disabled']:
        cat_path = os.path.join(top_path, cat)
        os.makedirs(cat_path, exist_ok=True)            
```

```python
train_size = 500
validation_size = 100
test_size = 100

stages_samples = {
    "train": train_size,
    "validation": validation_size,
    "test": test_size
}


import urllib.request 
from PIL import Image

def save_image(folder, _id, url):
    final_file = os.path.join(folder, _id + '.jpg')
    
    if os.path.isfile(final_file):
        print('.')
        return 
    
    with urllib.request.urlopen(url, timeout=2) as response:
        info = response.info()
        mime_type = info.get_content_type()
        
        ext = ''
        if mime_type == 'image/png':
            ext = 'png'
        elif mime_type == 'image/jpeg':
            ext = 'jpg'
            
        local_file = os.path.join(folder, _id + '.' + ext)        
            
        assert(ext)
        
        f = open(local_file, 'wb')
        f.write(response.read())
        f.close()
        
                
        im = Image.open(local_file)
        rgb_im = im.convert('RGB')
        rgb_im.save(final_file)
        
        print('saved {}'.format(final_file))
                
        
        

def save_images(dest_folder, data, start, size):
    c = 0
    for i, row in enumerate(data[start:, :]):
        _id = row[0]
        url = row[1]            
        try:
            save_image(dest_folder, _id, url)
            c = c + 1
        except Exception as e:
            print(str(e))
            
        if c >= size:
            return start + i + 1

            



np.random.seed(42)
for label in ['published', 'disabled']:
    label_data = df[df[label] == 1].values
    np.random.shuffle(label_data)

    start = 0
    for stage, total in stages_samples.items():
        pth = os.path.join(base_dir, stage, label)
        print('path {}, start: {}'.format(pth, start))
        start = save_images(pth, label_data, start, total)
            
            
```

```python
for folder in ['train', 'validation', 'test']:
    for cat in ['published', 'disabled']:
        print('data for {} {}: {}'.format(folder, cat, len(os.listdir(os.path.join(base_dir, folder, cat)))))
```

```python
train_published_files = os.listdir(os.path.join(base_dir, 'train', 'published'))
train_disabled_files = os.listdir(os.path.join(base_dir, 'train', 'disabled'))

validation_published_files = os.listdir(os.path.join(base_dir, 'validation', 'published'))
validation_disabled_files = os.listdir(os.path.join(base_dir, 'validation', 'disabled'))

test_published_files = os.listdir(os.path.join(base_dir, 'test', 'published'))
test_disabled_files = os.listdir(os.path.join(base_dir, 'test', 'disabled'))

dup = set(train_published_files) & set(validation_published_files)
for f in dup:
    os.remove(os.path.join(base_dir, 'validation', 'published', f))
    
dup = set(train_disabled_files) & set(validation_disabled_files)
for f in dup:
    os.remove(os.path.join(base_dir, 'validation', 'disabled', f))
    

dup = set(train_published_files) & set(test_published_files)
for f in dup:
    os.remove(os.path.join(base_dir, 'test', 'published', f))
    
dup = set(train_disabled_files) & set(test_disabled_files)
for f in dup:
    os.remove(os.path.join(base_dir, 'test', 'disabled', f))

    
dup = set(validation_published_files) & set(test_published_files)
for f in dup:
    os.remove(os.path.join(base_dir, 'test', 'published', f))
    
dup = set(validation_disabled_files) & set(test_disabled_files)
for f in dup:
    os.remove(os.path.join(base_dir, 'test', 'disabled', f))
    
```
